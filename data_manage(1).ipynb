```python
import pandas as pd
```

```python
def flatten(list1):
  return [item for sublist in list1 for item in sublist]
```

```python
df = pd.read_excel("Tabel Hero Power.xlsx", sheet_name="Sheet1")
all_team_name = [i for i in list(df.columns) if "Unnamed: " not in i]
all_num_player_team = [7, 8,8,10,6, 7, 6,8,8]
all_team_name2 = flatten([[all_team_name[i]]*all_num_player_team[i] for i in range(len(all_team_name))])
```
