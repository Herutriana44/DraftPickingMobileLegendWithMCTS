```python
import pandas as pd
import random

def random_float():
    return random.uniform(0.00, 1.00)
```

```python
df = pd.read_excel('DraftPickHeroML.xlsx')
```

```python
hero_col = list(df.columns)[2:]
```

```python
for hero in hero_col:
    df[hero] = [random_float() for i in range(len(df))]
```
